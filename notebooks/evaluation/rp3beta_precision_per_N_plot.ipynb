Here is the code for the plot used in R. Kwieciński, G. Melniczak and T. Górecki, "Comparison of Real-Time and Batch Job Recommendations," in IEEE Access, vol. 11, pp. 20553-20559, 2023, doi: 10.1109/ACCESS.2023.3249356.

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

df = pd.DataFrame([[200, 0.035121],
              [500, 0.035097],
              [10**5, 0.034999],
              [100, 0.034939],
              [300, 0.034890],
              [1000, 0.034866],
              [50, 0.034440],
              [40, 0.034051],
              [30, 0.033917],
              [20, 0.033698],
              [10, 0.031497],
            #   [5, 0.027350]
              ],
              columns=["N", "precision@10"]).sort_values(by='N').set_index('N')
ax = df.plot(logx=True, grid=True, ylabel="precision@10", legend=True)
ax.axvline(60, color="red", linestyle="--", label='N=60')
ax.legend(bbox_to_anchor=(1.00, 0.2))

images_dir = f"images"
if not os.path.isdir(images_dir):
    os.makedirs(images_dir)
plt.savefig(f"images/rt_precision_per_N.pdf", format="pdf")
```
